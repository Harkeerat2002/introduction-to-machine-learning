```python
import pandas as pd
import numpy as np
from sklearn import linear_model as lm
from sklearn.metrics import r2_score
```

```python
df = pd.read_csv('data/train.csv')
df.head()
```

```python
tf = pd.read_csv('data/test.csv')
tf.head()
```

```python
X = df[['x1', 'x2', 'x3', 'x4', 'x5', 'x6', 'x7', 'x8', 'x9', 'x10']]
y = df['y']

train_X = X[:80]
train_y = y[:80]

test_X = X[80:]
test_y = y[80:]
```

```python
regr = lm.LinearRegression()
regr.fit(train_X, train_y)
```

```python
r2 = r2_score(test_y, regr.predict(test_X))
print(r2)
```

```python
testData = tf[['x1', 'x2', 'x3', 'x4', 'x5', 'x6', 'x7', 'x8', 'x9', 'x10']]
testData.head()
```

```python
predictions = regr.predict(testData)
predictions
```

```python
id = tf['Id']
id.head()
```

```python
submission = pd.DataFrame({
        "Id": id,
        "y": predictions
    })
submission.to_csv('submission.csv', index=False)
submission.head()
```
